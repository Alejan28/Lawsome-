```python
import pandas as pd

CSV_PATH = "dataset_cereri_finale_categorizat.csv"
df = pd.read_csv(CSV_PATH, encoding="utf-8-sig")


print(df.head())


print(df.columns)
```

```python
import matplotlib.pyplot as plt


counts = df["ramura_dreptului"].value_counts()


plt.figure(figsize=(10,6))
counts.plot(kind="bar", color="skyblue")
plt.xlabel("Ramura dreptului")
plt.ylabel("Număr de cereri")
plt.title("Număr de cereri pe ramura dreptului")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = df["tip_actiune"].value_counts()


plt.figure(figsize=(12,6))
counts.plot(kind="bar", color="lightgreen")
plt.xlabel("Tip de acțiune")
plt.ylabel("Număr de cereri")
plt.title("Număr de cereri pe tip de acțiune")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

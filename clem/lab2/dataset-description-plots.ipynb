```python
from pathlib import Path
import matplotlib.pyplot as plt

DATA_DIR = Path("data")  # root directory containing your type folders

# collect counts
folders = []
counts = []

for folder in DATA_DIR.iterdir():
    if folder.is_dir():
        num_files = len(list(folder.glob("*.txt")))  # count .txt files
        folders.append(folder.name)
        counts.append(num_files)

# plot
plt.figure(figsize=(8,5))
plt.bar(folders, counts, color='skyblue')
plt.xlabel("Document Type")
plt.ylabel("Number of Files")
plt.title("Number of Files per Document Type")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
